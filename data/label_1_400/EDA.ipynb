```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 200->400 train

```python
labels = pd.read_csv("train_1_400_center_ord1.csv")
label_counts = labels.loc[labels['label'] < 400]['label'].value_counts().sort_index()
num_subplots = 6
data_chunks = np.array_split(label_counts, num_subplots)
fig, axs = plt.subplots(3,2, figsize=(25,  20))
for i, data_chunk in enumerate(data_chunks):
    row = i // 2
    col = i % 2
    
    # Vẽ biểu đồ cột
    data_chunk.plot(kind='bar', title=f'Biểu đồ {i + 1}',ax=axs[row, col])
    axs[row, col].set_xlabel('Label')
    axs[row, col].set_ylabel('Count')
```

```python
import pandas as pd

     

train_signer = []
test_signer = []
val_signer = []
full = pd.read_csv("train_1_400_center_ord1.csv",sep=',')
for idx,row in full.iterrows():
    name = row['name'].split('_center_ord1')[0].split('_')[-1]
    if name not in train_signer:
        train_signer.append(name)

full = pd.read_csv("val_1_400_center_ord1.csv",sep=',')
for idx,row in full.iterrows():
    name = row['name'].split('_center_ord1')[0].split('_')[-1]
    if name not in val_signer:
        val_signer.append(name)

full = pd.read_csv("test_1_400_center_ord1.csv",sep=',')
for idx,row in full.iterrows():
    name = row['name'].split('_center_ord1')[0].split('_')[-1]
    if name not in test_signer:
        test_signer.append(name)


# kiểm tra xem liệu có bị trung không
print([i for i in test_signer if i in train_signer]) # Không có test nằm trong train
print([i for i in test_signer if i in val_signer])  # không có test nằm trong val
print([i for i in train_signer if i in val_signer]) # không có train nằm trong val
len(train_signer),len(test_signer),len(val_signer),len(list(set(train_signer+ test_signer + val_signer)))
```
